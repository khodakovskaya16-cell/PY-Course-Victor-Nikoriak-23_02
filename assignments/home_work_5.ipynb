
# Task 1 — The Guessing Game.
 ==============================

```python
import random
number = random.randint(1, 10)
guess = int(input("Guess the number (1-10): "))
if guess == number:
    print("You guessed the number!")
else:
    print("You did not guess the number. The number was:", number)
```

# Task 2 — The birthday greeting program
 ==============================

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))
next_age = age + 1
print("Hello", name + ", on your next birthday you’ll be", next_age, "years")
```

# Task 3 — Words combination
==============================

```python
import random
word = input("Enter a word: ")
for i in range(5):
    random_word = ""
    for j in range(len(word)):
        random_word += random.choice(word)
    print(random_word)
```
