## Чому це важливо?

Як розробник, ви часто бачите, як розробники намагаються вимірювати швидкість коду в секундах. **Це пастка.** Секунди залежать від потужності процесора, фонових задач ОС та оперативної пам'яті.

**Нотація «О-велике» (Big O)** — це принципово інший інструмент. Вона не вимірює час у секундах, вона вимірює **масштабованість**. Вона відповідає на одне запитання:

> *«Як швидко зростатиме кількість операцій, якщо обсяг вхідних даних ($n$) збільшиться з 10 до 10 мільйонів?»*

## Частина 1 — Ментальна модель та Історія нотації

### Чому не секунди?

Один і той самий Python-скрипт виконається за 0.5 секунди на M3 MacBook і за 15 секунд на старому ноутбуці. Секунди — це характеристика **заліза**, а не **коду**.

Big O вимірює **кількість кроків (операцій)** відносно розміру вхідних даних $n$. Це апаратно-незалежна метрика.

### Походження нотації: від математики до CS

Нотація Big O (або «Ландау нотація») має математичне коріння, що сягає початку XX століття. Вона використовувалась у **теорії чисел** ще до того, як з'явились комп'ютери.

**Едмунд Ландау** (1877–1938) — німецький математик, після якого нотація отримала неофіційну назву «Ландау нотація».

**Дональд Кнут** — автор монументальної серії «The Art of Computer Programming» і батько формального аналізу алгоритмів. У 1976 році Кнут запропонував, щоб CS-спільнота прийняла $O$, $\Omega$, і $\Theta$ як стандартну математичну мову для обговорення швидкості алгоритмів.

### Чому Big O взагалі з'явилось?

На ранніх комп'ютерах ресурси були катастрофічно обмежені — ефективність алгоритмів мала пріоритет над усім. Але вимірювати час у секундах — фундаментально хибний підхід, бо результат залежить від:
- конкретного заліза (процесор, RAM)
- операційної системи та фонових процесів
- компілятора / інтерпретатора
- мови програмування

Потрібна була **апаратно-незалежна метрика**. Big O — це «золота середина»: математична точність без зайвих деталей реалізації. Вона дозволяє ігнорувати тактову частоту процесора і зосередитись виключно на тому, як ресурсоємність алгоритму зростає зі збільшенням $n$.

### Що таке $n$?

$n$ — це **розмір вхідних даних**. Залежно від контексту:
- кількість елементів у списку
- кількість рядків у файлі
- кількість вузлів у дереві
- кількість записів у базі даних

> **Головне запитання Big O:** «Якщо $n$ зросте у 10 разів, кількість операцій зросте у скільки разів?»

## Частина 2 — Усі Класи Складності

### Ментальна модель: Книжкова полиця

Уявіть книжкову полицю, де $n$ — це кількість книг:

| Клас | Назва | Аналогія з полицею | Python приклад |
|------|-------|-------------------|----------------|
| **O(1)** | Константний | *«Чи є хоч одна книга?»* — один погляд | `my_list[0]`, `my_dict[key]` |
| **O(log n)** | Логарифмічний | *«Знайти книгу на відсортованій полиці»* — ділимо навпіл | `bisect.bisect_left(sorted_list, x)` |
| **O(n)** | Лінійний | *«Прочитати всі книги»* — перебираємо кожну | `for item in my_list` |
| **O(n log n)** | Квазілінійний | *«Відсортувати книги»* — найкращий можливий сорт | `my_list.sort()` (Timsort) |
| **O(n²)** | Квадратичний | *«Знайти дублікати»* — кожну з кожною | Вкладені `for` цикли |
| **O(2ⁿ)** | Експоненційний | *«Всі можливі підбірки книг»* — кожна нова книга подвоює варіанти | Перебір усіх підмножин |
| **O(n!)** | Факторіальний | *«Всі порядки книг на полиці»* — кількість перестановок | Задача комівояжера |

### Таблиця зростання: скільки операцій при різних n?

| n | O(1) | O(log n) | O(n) | O(n log n) | O(n²) | O(2ⁿ) | O(n!) |
|---|------|----------|------|------------|-------|--------|-------|
| 1 | 1 | 0 | 1 | 0 | 1 | 2 | 1 |
| 10 | 1 | 3 | 10 | 33 | 100 | 1,024 | 3,628,800 |
| 100 | 1 | 7 | 100 | 664 | 10,000 | 10³⁰ | ∞ |
| 1,000 | 1 | 10 | 1,000 | 9,966 | 1,000,000 | ∞ | ∞ |
| 1,000,000 | 1 | 20 | 1,000,000 | 19,931,568 | 10¹² | ∞ | ∞ |

> **Ключовий висновок:** O(2ⁿ) та O(n!) є **практично непридатними** для n > 30-40 елементів. Жоден сучасний суперкомп'ютер не завершить такий розрахунок за розумний час.

### Практичний поріг застосовності

```
O(1)     → Будь-який n         ✅ Завжди безпечно
O(log n) → n до мільярдів      ✅ Масштабується ідеально
O(n)     → n до сотень мільйонів ✅ Прийнятно з оптимізацією
O(n log n) → n до десятків мільйонів ✅ Стандарт для сортування
O(n²)    → n до ~10,000        ⚠️ Небезпечно на великих даних
O(2ⁿ)   → n до ~30            ❌ Тільки для малих задач
O(n!)    → n до ~12            ❌ Практично непридатно
```

## Частина 3 — Архітектурний Довідник Патернів

Як розробник, ви можете проводити «миттєвий аналіз» складності коду, просто поглянувши на його **структурні патерни**. Big O — це інструмент розпізнавання патернів.

---

### Патерн 1: «Прямий доступ» — O(1) Константний час

**Ідея:** Виконання коду не залежить від розміру даних. Ви звертаєтесь до пам'яті безпосередньо.

**Як розпізнати:** У коді немає обходу даних (циклів) або рекурсії.

**Прояви у Python:**
- Звернення до елемента за індексом `my_list[i]`
- `len(my_list)` — Python не рахує елементи щоразу, а просто читає збережене значення з C-структури
- `my_list.append(value)` — амортизований O(1)
- `my_dict[key]`, `key in my_dict`, `item in my_set` — хешування = миттєва адреса

---

### Патерн 2: «Поділяй і Володарюй» — O(log n) Логарифмічний час

**Ідея:** На кожному кроці алгоритм **відкидає половину** даних, що залишилися. Надзвичайно ефективний патерн: для 1 мільярда елементів потрібно лише ~30 кроків.

**Як розпізнати:** Дані або область пошуку поділяються навпіл на кожній ітерації.

**Прояви у Python:**
- Бінарний пошук у відсортованих масивах
- Модуль `bisect` — архітектурно правильний спосіб бінарного пошуку в Python

---

### Патерн 3: «Конвеєр» — O(n) Лінійний час

**Ідея:** Програма змушена «торкнутися» кожного елемента даних хоча б один раз. Час виконання зростає пропорційно до кількості даних.

**Як розпізнати:** Один цикл `for` або `while` по всіх елементах.

**Приховані O(n) пастки у Python:**
- `value in my_list` — Python сканує список від початку до кінця
- `my_list.insert(0, value)` або `my_list.remove(value)` — зсуває всі n елементів у пам'яті
- `my_list.reverse()` — переставляє всі елементи

---

### Патерн 4: «Сортування» — O(n log n) Квазілінійний час

**Ідея:** Алгоритм розбиває дані на частини, сортує їх, а потім зливає разом. Це **еталонна нижня межа** для універсальних алгоритмів сортування — швидше, ніж O(n log n), загального сортування не існує.

**Як розпізнати:** Операція «поділяй і володарюй» виконується по одному разу для кожного елемента.

**Прояви у Python:**
- `my_list.sort()` та `sorted(iterable)` — Python використовує **Timsort** (гібрид MergeSort + InsertionSort)

---

### Патерн 5: «Декартовий вибух» — O(n²) Квадратичний час

**Ідея:** Подвійна робота. Для кожного елемента колекції ми перевіряємо кожен інший елемент. На великих обсягах даних цей патерн призводить до повної зупинки системи.

**Як розпізнати:** Два вкладених цикли, кожен з яких перебирає дані розміром n.

**Прояви у Python:**
- Пошук дублікатів без `set`: зовнішній цикл × внутрішній = n × n кроків
- Матричні обчислення з двома вкладеними циклами

> **Критичне правило:** Виклик функції не є O(1)! Він включає **загальну кількість кроків всередині функції**. Якщо ви викликаєте `my_list.index(x)` (O(n)) всередині циклу (O(n)) — ви непомітно будуєте O(n²) систему.

---

### Патерн 6: «Комбінаторний колапс» — O(2ⁿ) та O(n!)

**O(2ⁿ) — Експоненційний:**
- Кожен новий елемент **подвоює** час роботи
- Типовий сценарій: перебір всіх можливих підмножин
- Практично придатний лише для n < 30

**O(n!) — Факторіальний:**
- Перебір всіх можливих **перестановок**
- Класичний приклад: задача комівояжера (знайти найкоротший шлях через усі міста)
- Для n=12: 479,001,600 операцій. Для n=20: ~2.4 квінтильйони операцій

```python
# O(2^n): генерація всіх підмножин (n=20 → 1,048,576 підмножин)
def all_subsets(items):
    if not items:
        return [[]]
    first = items[0]
    rest_subsets = all_subsets(items[1:])           # Рекурсія на n-1
    return rest_subsets + [[first] + s for s in rest_subsets]  # Подвоюємо!

# O(n!): генерація всіх перестановок (n=10 → 3,628,800 перестановок)
from itertools import permutations
all_orders = list(permutations(range(10)))  # Ніколи не робіть це для n>12!
```

````python
# ===================================================
# ДЕМОНСТРАЦІЯ: Кожен клас складності на практиці
# Запускайте кожну функцію і спостерігайте за часом
# ===================================================
import time
import random

# Генеруємо список із 100,000 елементів для тестів
DATA = list(range(100_000))
DATA_SET = set(DATA)      # Та сама інформація, але у формі множини
DATA_DICT = {k: k for k in DATA}  # Та сама інформація у формі словника

print(f"Розмір тестових даних: {len(DATA):,} елементів")
print("-" * 50)

# --------------------------------------------------
# O(1) — Константний час
# Кількість операцій НЕ залежить від розміру даних
# --------------------------------------------------
def o_constant_list(data):
    """Доступ до елемента списку за індексом — завжди 1 крок."""
    return data[0]  # Обчислення зміщення пам'яті: address = base + 0 * sizeof(ptr)

def o_constant_dict(data_dict, key):
    """Пошук у словнику за ключем — хешування, завжди ~1 крок."""
    return data_dict.get(key)  # hash(key) → адреса комірки → значення

# --------------------------------------------------
# O(log n) — Логарифмічний час
# Кожен крок ВДВІЧІ зменшує область пошуку
# --------------------------------------------------
import bisect

def o_logarithmic(sorted_data, target):
    """Бінарний пошук: для 100,000 елементів — максимум 17 кроків."""
    # log2(100_000) ≈ 17
    # Бібліотека bisect реалізує бінарний пошук на відсортованому списку
    index = bisect.bisect_left(sorted_data, target)
    if index < len(sorted_data) and sorted_data[index] == target:
        return index
    return -1

# --------------------------------------------------
# O(n) — Лінійний час
# Кожен елемент перевіряється рівно один раз
# --------------------------------------------------
def o_linear_search(data, target):
    """Лінійний пошук: у найгіршому випадку переберемо всі n елементів."""
    for i, item in enumerate(data):  # i — лічильник кроків
        if item == target:
            return i
    return -1  # Елемент не знайдений — пройшли весь список

# --------------------------------------------------
# O(n log n) — Квазілінійний час
# Python використовує алгоритм Timsort (гібрид MergeSort + InsertionSort)
# --------------------------------------------------
def o_n_log_n(data):
    """Сортування — абсолютна нижня межа для алгоритмів порівняння."""
    return sorted(data)  # Timsort: O(n log n) у середньому та найгіршому випадку

# --------------------------------------------------
# O(n²) — Квадратичний час
# Для кожного елемента ми проходимо по ВСІХ елементах
# --------------------------------------------------
def o_quadratic_find_duplicates(data):
    """Пошук дублікатів через вкладені цикли — класична O(n²) пастка."""
    duplicates = []
    for i in range(len(data)):          # Зовнішній цикл: n ітерацій
        for j in range(i + 1, len(data)):  # Внутрішній цикл: (n-i-1) ітерацій
            if data[i] == data[j]:      # Порівняння: n*(n-1)/2 ≈ n² разів
                duplicates.append(data[i])
    return duplicates

# --------------------------------------------------
# Виміряємо час виконання на малих даних
# --------------------------------------------------
small_data = list(range(1000))  # 1000 елементів для демонстрації
random.shuffle(small_data)      # Перемішуємо, щоб зробити пошук реалістичним

operations = [
    ("O(1)  — доступ за індексом", lambda: o_constant_list(small_data)),
    ("O(1)  — пошук у словнику",   lambda: o_constant_dict(DATA_DICT, 500)),
    ("O(log n) — бінарний пошук",  lambda: o_logarithmic(DATA, 99_999)),
    ("O(n)  — лінійний пошук",     lambda: o_linear_search(small_data, -1)),
    ("O(n log n) — сортування",    lambda: o_n_log_n(small_data)),
]

for name, func in operations:
    start = time.perf_counter()
    for _ in range(1000):  # Повторюємо 1000 разів для точності вимірювання
        func()
    elapsed = (time.perf_counter() - start) / 1000 * 1_000_000  # мікросекунди
    print(f"{name}: {elapsed:.2f} мкс")

print("\n[Примітка] O(n²) не запускаємо — на 1000 елементах це 1,000,000 операцій,")
print("           на 10,000 — вже 100,000,000. Дивіться наступну клітинку.")
````

````python
# ===================================================
# ПОРІВНЯННЯ МАСШТАБУВАННЯ: Як зростає час при n × 10
# Це головна суть Big O — поведінка при збільшенні n
# ===================================================
import time

def measure_time(func, *args):
    """Вимірює час виконання функції у мікросекундах."""
    start = time.perf_counter()
    func(*args)
    return (time.perf_counter() - start) * 1_000_000

# Функції для тестування масштабування
def linear_sum(n):
    """O(n): підсумовуємо числа від 0 до n."""
    total = 0
    for i in range(n):  # n ітерацій
        total += i
    return total

def quadratic_pairs(n):
    """O(n²): знаходимо всі пари чисел від 0 до n."""
    count = 0
    for i in range(n):       # n ітерацій
        for j in range(n):   # ще n ітерацій для КОЖНОЇ зовнішньої
            count += 1
    return count

# Тестуємо при різних значеннях n
print(f"{'n':>10} | {'O(n) мкс':>12} | {'O(n²) мкс':>12} | {'Різниця':>10}")
print("-" * 55)

for n in [100, 1_000, 2_000, 5_000]:
    t_linear    = measure_time(linear_sum, n)
    t_quadratic = measure_time(quadratic_pairs, n)
    ratio = t_quadratic / t_linear if t_linear > 0 else 0
    print(f"{n:>10,} | {t_linear:>12.1f} | {t_quadratic:>12.1f} | {ratio:>9.0f}x")

print("\n[Висновок]")
print("При n=1,000 → O(n²) повільніший за O(n) приблизно в 1,000 разів.")
print("При n=10,000 → різниця стане ~10,000 разів!")
print("Ось чому O(n²) вбиває продакшн-системи при масштабуванні.")
````

# Частина 3 — Три базові правила аналізу Big O

Big O потрібен не для того, щоб порахувати точний час у секундах.
Він потрібен для відповіді на головне питання:

> **Як буде зростати обсяг роботи алгоритму, якщо даних стане більше?**

Тобто Big O описує не “швидкість комп’ютера”, а **масштабування алгоритму**.

---

## Перед тим як перейти до правил: що таке `n`

У Big O буква `n` означає:

> **розмір вхідних даних**

Наприклад:

* якщо функція працює зі списком із 10 елементів, то `n = 10`
* якщо список має 1 000 елементів, то `n = 1000`

Якщо даних стає більше, ми дивимося, як змінюється кількість дій.

---

# Правило 1. Аналізуємо найгірший випадок

## Суть правила

Коли ми визначаємо Big O, ми беремо **найгірший сценарій виконання алгоритму**.

Чому?
Тому що в реальних системах важливо знати, що буде не тоді, коли “пощастило”, а тоді, коли дані розташовані невигідно.

---

## Приклад

```python
def find_user(users, target_id):
    for user in users:
        if user.id == target_id:
            return user
    return None
```

---

## Як тут рахувати

Нехай у списку `users` є `n` користувачів.

Цикл:

```python
for user in users:
```

може пройти:

* 1 раз, якщо елемент перший
* 5 разів, якщо елемент п’ятий
* `n` разів, якщо елемент останній
* `n` разів, якщо елемента немає взагалі

---

## Чому беремо саме найгірший випадок

Тому що Big O відповідає на питання:

> **Скільки максимально роботи може виконати алгоритм при зростанні даних?**

У цьому прикладі максимум — перевірити всі елементи.

---

## Висновок

```text
O(n)
```

---

## Просте правило мислення

Якщо у вас є **один цикл по всьому списку**, то найчастіше це:

```text
O(n)
```

## Ще один приклад

```python

        if number == target:
            return True
    return False
```

Тут логіка така сама:

* у найгіршому випадку перевіряємо всі `n` елементів
* отже:

```text
O(n)
```

# Правило 2. Константи не впливають на порядок складності

## Суть правила

Big O не цікавиться точним числом дрібних дій.
Його цікавить **тип росту**.

Тобто:

* `n`
* `2n`
* `5n`

у Big O записуються однаково:

```text
O(n)
```

Бо всі вони ростуть **лінійно**.

---

## Приклад 1 — два окремі цикли

```python
def process_data(data):
    for x in data:
        print(x)

    for x in data:
        print(x * 2)
```

---

## Як рахувати

Нехай у `data` є `n` елементів.

Перший цикл виконується `n` разів.
Другий цикл теж виконується `n` разів.

Разом:

```text
n + n = 2n
```

---

## Який буде Big O

У Big O це записують як:

```text
O(n)
```

---

## Чому не `O(2n)`

Тому що коефіцієнт 2 не змінює сам характер росту.

Порівняй:

* при `n = 10` → `2n = 20`
* при `n = 1000` → `2n = 2000`
* при `n = 1 000 000` → `2n = 2 000 000`

Це все одно **лінійне зростання**.

## Суть правила

## Частина 4б — Big Omega (Ω) та Big Theta (Θ): Повна Картина

Big O — найвідоміша, але не єдина асимптотична нотація. Теорія складності використовує три нотації для повного математичного опису алгоритму.

---

### Big O (O) — Верхня межа: «Не гірше, ніж...»

Big O описує **найгірший сценарій** — верхню межу часу виконання. Це гарантія: «алгоритм виконається не більше, ніж за X кроків».

```python
def linear_search(data, target):
    for item in data:        # У найгіршому випадку — всі n елементів
        if item == target:
            return True
    return False
# Big O: O(n) — гарантія верхньої межі
```

---

### Big Omega (Ω) — Нижня межа: «Не краще, ніж...»

Big Ω описує **найкращий сценарій** — нижню межу. Це мінімальна кількість кроків, яку алгоритм виконає **завжди**, навіть при ідеальних умовах.

```python
def linear_search(data, target):
    for item in data:
        if item == target:
            return True   # Якщо target — перший елемент → 1 крок
    return False
# Big Ω: Ω(1) — у найкращому випадку (target є першим елементом)
```

---

### Big Theta (Θ) — Точна межа: «Рівно...»

Big Θ використовується, коли верхня та нижня межі **збігаються**. Це найточніша характеристика — алгоритм завжди виконається рівно за Θ(f(n)) кроків.

```python
def sum_all(data):
    total = 0
    for item in data:   # Завжди рівно n ітерацій — жодних скорочень
        total += item
    return total
# Big Θ: Θ(n) — і найкращий, і найгірший сценарій однакові
# Big O: O(n) ✓  |  Big Ω: Ω(n) ✓  |  Big Θ: Θ(n) ✓
```

---

### Зведена таблиця нотацій

| Нотація | Читається як | Описує | Питання |
|---------|-------------|--------|---------|
| **O(f(n))** | «О від f(n)» | Верхня межа (worst case) | «Найгірше, що може статись?» |
| **Ω(f(n))** | «Омега від f(n)» | Нижня межа (best case) | «Найкраще, що може статись?» |
| **Θ(f(n))** | «Тета від f(n)» | Точна межа (tight bound) | «Алгоритм завжди так поводиться?» |

### Приклад: лінійний пошук

| Нотація | Значення | Коли |
|---------|----------|------|
| O(n) | Верхня межа | Елемент в кінці або відсутній |
| Ω(1) | Нижня межа | Елемент — перший у списку |
| Θ — немає | — | O ≠ Ω, тому Θ не існує |

> **У практиці (та на співбесідах)** переважно використовують **Big O** — бо нас цікавить **гарантія продуктивності**, а не надія на щасливий збіг обставин.

````python
# ===================================================
# ПРАВИЛА АНАЛІЗУ: Демонстрація на Python-коді
# ===================================================

# -------------------------------------------
# ПРАВИЛО 2: Константи ігноруються
# -------------------------------------------
def two_passes(data):
    """
    Два незалежні цикли по одному масиву.
    Фактична кількість операцій: 2n
    Big O запис: O(n) — константа 2 відкидається.
    """
    # Перший прохід: знаходимо максимум (n операцій)
    maximum = data[0]
    for item in data:
        if item > maximum:
            maximum = item

    # Другий прохід: знаходимо мінімум (ще n операцій)
    minimum = data[0]
    for item in data:
        if item < minimum:
            minimum = item

    # Загальна кількість операцій: 2n
    # Але Big O: O(n), тому що ми дивимось на ПОВЕДІНКУ, а не точну формулу
    return minimum, maximum

# -------------------------------------------
# ПРАВИЛО 3: Домінування вищих порядків
# -------------------------------------------
def mixed_complexity(data):
    """
    Функція з кількома блоками різної складності.
    Big O визначається ДОМІНУЮЧИМ доданком.
    """
    n = len(data)

    # Блок 1: O(1) — константний час (знаходимо перший елемент)
    first = data[0]                # 1 операція

    # Блок 2: O(n) — лінійний час (знаходимо суму)
    total = sum(data)              # n операцій

    # Блок 3: O(n²) — квадратичний час (знаходимо всі пари)
    pairs_count = 0
    for i in range(n):             # n ітерацій
        for j in range(i + 1, n): # (n-i-1) ітерацій
            pairs_count += 1       # Разом: n*(n-1)/2 ≈ n²/2 операцій

    # Загальна формула: O(1) + O(n) + O(n²/2)
    # Скорочуємо: відкидаємо константи та нижчі порядки
    # РЕЗУЛЬТАТ: O(n²) — домінуючий член
    return first, total, pairs_count

# Демонстрація
test_data = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
print("two_passes([3,1,4,...]):", two_passes(test_data))
print("\nАналіз складності mixed_complexity():")
print("  Блок 1 (перший елемент): O(1)")
print("  Блок 2 (sum):            O(n)")
print("  Блок 3 (вкладені цикли): O(n²)")
print("  Загальний Big O:         O(n²) — домінує квадратичний блок")
````

## Частина 5 — Реалізації: O(n), O(log n), O(n²) у Python

---

### O(n): Лінійний пошук — Патерн «Конвеєр»

**Концепція:** Алгоритм є O(n), якщо кількість операцій зростає **прямо пропорційно** до розміру вхідних даних. Якщо обсяг даних подвоюється — час виконання теж подвоюється.

**Як розпізнати в коді:** Один цикл, який перебирає структуру даних.

```python
def linear_search(data, target):
    # O(n) — у найгіршому випадку перевіряємо кожен елемент
    for index, value in enumerate(data):
        if value == target:
            return index
    return -1

# Приховані O(n) у Python — знайте їх напам'ять:
# min(data)         → O(n)
# max(data)         → O(n)
# sum(data)         → O(n)
# value in my_list  → O(n) — лінійне сканування!
# my_list.index(x)  → O(n) — НЕ O(1)!
# my_list.count(x)  → O(n)
```

---

### O(n²): Квадратичний пошук — Патерн «Декартовий вибух»

**Концепція:** Вкладені цикли, де обидва залежать від n. Подвоєння даних → учетверення часу.

```python
def has_duplicates_naive(data):
    # O(n²): для кожного з n елементів — ще один прохід по n елементах
    n = len(data)
    for i in range(n):           # Зовнішній: n ітерацій
        for j in range(i + 1, n):  # Внутрішній: до (n-1) ітерацій
            if data[i] == data[j]:
                return True
    return False
# Загальна кількість порівнянь: n*(n-1)/2  ≈  n²/2  →  O(n²)
```

---

### O(log n): Бінарний пошук — Патерн «Поділяй і Володарюй»

**Концепція:** Кожен крок **вдвічі** зменшує область пошуку. Вимагає **відсортованих** даних.

**Як це працює:**
1. Визначаємо вікно пошуку через вказівники `low` і `high`
2. Стрибаємо до середнього елемента `mid`
3. Якщо `target < sorted_data[mid]` → відкидаємо ПРАВУ половину (переміщуємо `high`)
4. Якщо `target > sorted_data[mid]` → відкидаємо ЛІВУ половину (переміщуємо `low`)
5. Повторюємо до знаходження або `low > high`

```python
def binary_search_manual(sorted_data, target):
    # O(log n): кожна ітерація відкидає половину залишкових елементів
    low, high = 0, len(sorted_data) - 1

    while low <= high:
        mid = (low + high) // 2   # Знаходимо центр поточного вікна

        if sorted_data[mid] == target:
            return mid             # Знайшли!
        elif target < sorted_data[mid]:
            high = mid - 1        # Відкидаємо праву половину
        else:
            low = mid + 1         # Відкидаємо ліву половину

    return -1  # Елемент не знайдений
```

> **Увага:** Написання бінарного пошуку вручну — класика для співбесід, але в реальному коді **завжди** використовуйте `bisect`. Дослідження показали, що ~90% програмістів допускають помилки (off-by-one, integer overflow) при ручній реалізації.

```python
import bisect

def binary_search_pythonic(sorted_data, target):
    # O(log n): Python C-реалізація — оптимізована і безпомилкова
    index = bisect.bisect_left(sorted_data, target)
    # bisect_left повертає позицію, де target МАЄ бути вставлений зліва
    # Перевіряємо, чи елемент на цій позиції дійсно є target
    if index < len(sorted_data) and sorted_data[index] == target:
        return index
    return -1
```

````python
# ===================================================
# РЕАЛІЗАЦІЇ: Лінійний пошук, бінарний пошук (вручну + bisect), O(n²)
# ===================================================
import bisect
import time
import math

# -------------------------------------------
# O(n): Лінійний пошук
# -------------------------------------------
def linear_search(data, target):
    """
    O(n) — перевіряємо кожен елемент по черзі.
    Найгірший сценарій: target відсутній → n порівнянь.
    """
    for index, value in enumerate(data):
        if value == target:
            return index
    return -1

# -------------------------------------------
# O(n²): Пошук дублікатів через вкладені цикли
# -------------------------------------------
def has_duplicates_naive(data):
    """
    O(n²) — класична пастка з вкладеними циклами.
    n*(n-1)/2 порівнянь у найгіршому випадку.
    """
    n = len(data)
    for i in range(n):            # Зовнішній: n ітерацій
        for j in range(i + 1, n): # Внутрішній: (n-i-1) ітерацій
            if data[i] == data[j]:
                return True
    return False

# -------------------------------------------
# O(log n): Бінарний пошук — ручна реалізація
# -------------------------------------------
def binary_search_manual(sorted_data, target):
    """
    O(log n) — кожна ітерація відкидає половину простору пошуку.
    Крок 1: low=0, high=n-1
    Крок 2: mid = (low+high)//2 → порівнюємо
    Крок 3: переміщуємо low або high → відкидаємо половину
    Повторюємо до знаходження або low > high
    """
    low, high = 0, len(sorted_data) - 1
    steps = 0  # Лічимо кроки для наочності

    while low <= high:
        steps += 1
        mid = (low + high) // 2   # Центр поточного вікна

        if sorted_data[mid] == target:
            return mid, steps      # Знайшли! Повертаємо позицію та кількість кроків
        elif target < sorted_data[mid]:
            high = mid - 1        # Відкидаємо праву половину
        else:
            low = mid + 1         # Відкидаємо ліву половину

    return -1, steps

# -------------------------------------------
# O(log n): Бінарний пошук — через bisect (Pythonic)
# -------------------------------------------
def binary_search_bisect(sorted_data, target):
    """
    O(log n) — C-рівнева реалізація Python стандартної бібліотеки.
    bisect_left повертає позицію вставки: перший індекс >= target.
    """
    index = bisect.bisect_left(sorted_data, target)
    if index < len(sorted_data) and sorted_data[index] == target:
        return index
    return -1

# -------------------------------------------
# Демонстрація бінарного пошуку — покроково
# -------------------------------------------
print("=== Покрокова демонстрація бінарного пошуку ===")
print()

sorted_list = list(range(0, 100, 2))  # [0, 2, 4, 6, ..., 98] — 50 елементів
target = 72

print(f"Список: [0, 2, 4, ..., 98] — {len(sorted_list)} елементів")
print(f"Шукаємо: {target}")
print(f"Теоретична максимальна кількість кроків: {math.ceil(math.log2(len(sorted_list)))}")
print()

index, steps = binary_search_manual(sorted_list, target)
print(f"Знайдено на позиції: {index}  |  Кроків: {steps}")
print(f"Перевірка: sorted_list[{index}] = {sorted_list[index] if index != -1 else 'N/A'}")

# -------------------------------------------
# Порівняння: лінійний vs бінарний на великих даних
# -------------------------------------------
print("\n=== Порівняння O(n) vs O(log n) ===")
print()

big_sorted = list(range(0, 2_000_000, 2))  # 1 мільйон парних чисел
target_val = 1_999_998  # Майже кінець — найгірший сценарій для лінійного

# Лінійний: O(n)
start = time.perf_counter()
for _ in range(100):
    linear_search(big_sorted, target_val)
t_linear = (time.perf_counter() - start) / 100 * 1000

# Бінарний (bisect): O(log n)
start = time.perf_counter()
for _ in range(100):
    binary_search_bisect(big_sorted, target_val)
t_binary = (time.perf_counter() - start) / 100 * 1000

n = len(big_sorted)
print(f"n = {n:,} елементів")
print(f"Лінійний пошук O(n):          {t_linear:.2f} мс  (до {n:,} порівнянь)")
print(f"Бінарний пошук O(log n):       {t_binary:.4f} мс  (до {math.ceil(math.log2(n))} порівнянь)")
print(f"Прискорення бінарного:         {t_linear/t_binary:.0f}x")
print()

# -------------------------------------------
# Демонстрація O(2^n) зростання (лише показуємо числа!)
# -------------------------------------------
print("=== Масштаб O(2^n) та O(n!) — чому вони непридатні ===")
print()
print(f"{'n':>5} | {'O(n)':>15} | {'O(n²)':>15} | {'O(2^n)':>20} | {'O(n!)':>25}")
print("-" * 90)
import math
for n in [5, 10, 15, 20, 25]:
    on = n
    on2 = n**2
    o2n = 2**n
    onf = math.factorial(n)
    # Форматуємо великі числа
    o2n_str = f"{o2n:,}" if o2n < 10**15 else f"~{o2n:.2e}"
    onf_str = f"{onf:,}" if onf < 10**15 else f"~{onf:.2e}"
    print(f"{n:>5} | {on:>15,} | {on2:>15,} | {o2n_str:>20} | {onf_str:>25}")
````

## Частина 6 — Python під капотом: list, dict, set

Python абстрагує управління пам'яттю, що робить дуже легким написання O(n²) алгоритмів без усвідомлення цього. Ви повинні знати, що відбувається всередині.

---

### Зведена таблиця часової складності

| Структура | Пошук (Lookup) | Вставка (Insert) | Видалення (Delete) | Ітерація |
|-----------|----------------|------------------|--------------------|---------|
| **`list`** | O(1) за індексом / **O(n) за значенням** | O(1) `append` / **O(n) `insert`** | O(1) `pop()` / **O(n) `pop(0)`** | O(n) |
| **`dict`** | **O(1)** за ключем | **O(1)** | **O(1)** | O(n) |
| **`set`** | **O(1)** `in` | **O(1)** `add` | **O(1)** `remove` | O(n) |

> **Примітка:** Для `dict` та `set` вказана середня складність. У найгіршому випадку (хеш-колізії) — O(n), але на практиці Python мінімізує такі ситуації.

---

### `list`: Безперервний масив посилань

Python `list` — це **не зв'язний список** (linked list)! Це безперервний масив посилань на об'єкти в пам'яті (contiguous array of object references).

```
Пам'ять:
[ptr→3] [ptr→1] [ptr→4] [ptr→1] [ptr→5]
  [0]     [1]     [2]     [3]     [4]
```

- `list[2]` → Python обчислює `base_address + 2 * 8 bytes` → **O(1)** ✅
- `insert(0, x)` → потрібно зсунути ВСІ елементи вправо на 1 позицію → **O(n)** ❌
- `append(x)` → Python виділяє пам'ять з запасом (overallocation). Амортизовано **O(1)** ✅

---

### `dict` та `set`: Хеш-таблиці

Словник і множина — це **хеш-таблиці**. Замість пошуку за позицією, вони обчислюють адресу через функцію `hash()`.

```
hash("python") → 8765432  → комірка №17 у таблиці → значення
```

- Lookup `dict[key]` → `hash(key)` → адреса → **O(1)** ✅
- Lookup `item in set` → `hash(item)` → перевірка комірки → **O(1)** ✅
- Пам'ять: словники споживають більше RAM, ніж списки (таблиця має порожні комірки для уникнення колізій)

> **З Python 3.7+** словники гарантовано зберігають порядок вставки ключів.

````python
# ===================================================
# ВНУТРІШНЯ АРХІТЕКТУРА: list vs dict vs set
# Вимірюємо реальну різницю продуктивності
# ===================================================
import time
import sys

# -------------------------------------------
# 1. Пошук елемента: list vs set vs dict
# -------------------------------------------
N = 1_000_000  # 1 мільйон елементів

# Створюємо три структури з однаковими даними
data_list = list(range(N))     # O(n) для пошуку
data_set  = set(range(N))      # O(1) для пошуку
data_dict = dict.fromkeys(range(N))  # O(1) для пошуку

target = N - 1  # Найгірший сценарій: шукаємо ОСТАННІЙ елемент (або елемент, якого немає)

# Вимірюємо час пошуку для кожної структури
def time_lookup(collection, target, iterations=1000):
    """Вимірює середній час однієї операції пошуку."""
    start = time.perf_counter()
    for _ in range(iterations):
        _ = target in collection  # Операція 'in'
    elapsed = (time.perf_counter() - start) / iterations * 1_000_000
    return elapsed

t_list = time_lookup(data_list, target)
t_set  = time_lookup(data_set, target)
t_dict = time_lookup(data_dict, target)

print("=== Пошук елемента (N = 1,000,000) ===")
print(f"  list:  {t_list:>10.2f} мкс  → O(n) — скануємо до 1,000,000 елементів")
print(f"  set:   {t_set:>10.4f} мкс  → O(1) — миттєво через хеш")
print(f"  dict:  {t_dict:>10.4f} мкс  → O(1) — миттєво через хеш")
print(f"  Різниця list/set: {t_list/t_set:.0f}x")

# -------------------------------------------
# 2. Вставка: append vs insert(0,...)
# -------------------------------------------
print("\n=== Вставка в список (N = 100,000) ===")

# O(1): append додає в КІНЕЦЬ — жодного зсуву
def test_append(n):
    lst = []
    for i in range(n):
        lst.append(i)  # Амортизований O(1): Python виділяє пам'ять із запасом
    return lst

# O(n): insert(0,...) вставляє на ПОЧАТОК — зсуває ВСЕ вправо
def test_insert_front(n):
    lst = []
    for i in range(n):
        lst.insert(0, i)  # O(n): кожна вставка зсуває всі попередні елементи
    return lst

n = 10_000

start = time.perf_counter()
test_append(n)
t_append = (time.perf_counter() - start) * 1000

start = time.perf_counter()
test_insert_front(n)
t_insert = (time.perf_counter() - start) * 1000

print(f"  append() × {n:,}:     {t_append:.2f} мс  → O(1) амортизований")
print(f"  insert(0,...) × {n:,}: {t_insert:.2f} мс  → O(n) катастрофа")
print(f"  Різниця: {t_insert/t_append:.0f}x")
print("  Рішення: використовуйте collections.deque для черг (O(1) з обох кінців)!")

# -------------------------------------------
# 3. Пам'ять: list vs dict
# -------------------------------------------
print("\n=== Витрати пам'яті (N = 1,000) ===")
small_list = list(range(1000))
small_dict = dict.fromkeys(range(1000))
small_set  = set(range(1000))

print(f"  list (1000 int):  {sys.getsizeof(small_list):,} байт")
print(f"  set  (1000 int):  {sys.getsizeof(small_set):,} байт")
print(f"  dict (1000 int):  {sys.getsizeof(small_dict):,} байт")
print("  → dict/set споживають ~2x більше пам'яті ніж list (хеш-таблиця потребує порожніх комірок)")
````

## Частина 7 — Антипатерни та Архітектурні Виправлення

Як розробник ви повинні розпізнавати ці пастки і виправляти їх на рівні структур даних — **до** написання логіки.

---

### Антипатерн №1: Пастка лінійного пошуку в циклі

Це найпоширеніша причина деградації сервісів.

```python
# АНТИПАТЕРН: O(n × m) — прихована квадратична бомба
def find_common_items_bad(list_a, list_b):
    result = []
    for item in list_a:        # O(n) — зовнішній цикл
        if item in list_b:     # O(m) — Python сканує ВЕСЬ list_b!
            result.append(item)
    # Загальна складність: O(n * m)
    # Якщо n=m=10,000 → 100,000,000 операцій
```

**Виправлення:** Перетворити `list_b` на `set` перед циклом.
```python
# ВИПРАВЛЕННЯ: O(n + m)
def find_common_items_good(list_a, list_b):
    set_b = set(list_b)         # O(m) — одноразова побудова хеш-таблиці
    return [item for item in list_a if item in set_b]  # O(n) × O(1)
    # Загальна складність: O(n + m)
```

---

### Антипатерн №2: `insert(0,...)` замість `deque`

Якщо вам потрібна черга (FIFO — first in, first out):
```python
# АНТИПАТЕРН: O(n) для кожної операції
queue = []
queue.insert(0, "task")  # O(n): зсуває всі елементи!
queue.pop(0)             # O(n): зсуває всі елементи!

# ВИПРАВЛЕННЯ: O(1) для обох операцій
from collections import deque
queue = deque()
queue.appendleft("task")  # O(1): справжній двобічний зв'язний список
queue.popleft()           # O(1): немає жодного зсуву
```

---

### Антипатерн №3: Ігнорування `bisect` для відсортованих списків

```python
# АНТИПАТЕРН: O(n) пошук у відсортованих даних
if target in sorted_list:   # Python сканує лінійно, ігноруючи порядок!
    ...

# ВИПРАВЛЕННЯ: O(log n) — бінарний пошук
import bisect
index = bisect.bisect_left(sorted_list, target)
found = index < len(sorted_list) and sorted_list[index] == target
```

---

### Коли Big O вас обманює: Амортизація та малі дані

**Принцип Роба Пайка:** *«Хитромудрі алгоритми повільно працюють при малих n, а n зазвичай маленьке».*

Якщо ви обробляєте 15 елементів — не ускладнюйте. O(n²) на 15 елементах = 225 операцій. Накладні витрати на `set()` можуть бути більшими.

**Амортизована складність:** `list.append()` — у найгіршому випадку O(n) (коли Python копіює масив), але в **середньому** O(1), тому що Python виділяє пам'ять із геометричним запасом (×1.125 при кожному розширенні).

````python
# ===================================================
# АНТИПАТЕРНИ: Порівняння поганого і правильного коду
# ===================================================
import time
from collections import deque
import bisect

# -------------------------------------------
# АНТИПАТЕРН №1: O(n²) vs O(n+m)
# Пошук спільних елементів двох списків
# -------------------------------------------
def find_common_bad(list_a, list_b):
    """O(n * m): для кожного елемента list_a — лінійний пошук у list_b."""
    result = []
    for item in list_a:      # Зовнішній цикл: n ітерацій
        if item in list_b:   # 'in' для list — O(m): сканує весь list_b!
            result.append(item)
    return result

def find_common_good(list_a, list_b):
    """O(n + m): будуємо хеш-таблицю один раз, потім O(1) lookup у циклі."""
    # Крок 1: O(m) — конвертуємо list_b у set (хеш-таблицю)
    set_b = set(list_b)
    # Крок 2: O(n) × O(1) — ітерація по list_a з O(1) перевіркою
    return [item for item in list_a if item in set_b]

# Тестові дані: два списки по 5,000 елементів з частковим перетином
list_a = list(range(5_000))
list_b = list(range(2_500, 7_500))  # Перетин: 2500–4999

start = time.perf_counter()
result_bad = find_common_bad(list_a, list_b)
t_bad = (time.perf_counter() - start) * 1000

start = time.perf_counter()
result_good = find_common_good(list_a, list_b)
t_good = (time.perf_counter() - start) * 1000

print("=== Антипатерн №1: Пошук спільних елементів (n=m=5,000) ===")
print(f"  ПОГАНО  (O(n*m)): {t_bad:.2f} мс  | Знайдено: {len(result_bad)} елементів")
print(f"  ДОБРЕ   (O(n+m)): {t_good:.4f} мс  | Знайдено: {len(result_good)} елементів")
print(f"  Прискорення: {t_bad/t_good:.0f}x")

# -------------------------------------------
# АНТИПАТЕРН №2: list як черга vs deque
# -------------------------------------------
print("\n=== Антипатерн №2: Черга — list.insert(0,...) vs deque ===")

n = 5_000  # Кількість операцій вставки

# Список: кожна вставка на початок зсуває всі елементи → O(n) на кожну операцію
start = time.perf_counter()
bad_queue = []
for i in range(n):
    bad_queue.insert(0, i)  # O(n): зсуває ВСЕ!
t_list_queue = (time.perf_counter() - start) * 1000

# deque: O(1) на кожну операцію — справжній двобічний список
start = time.perf_counter()
good_queue = deque()
for i in range(n):
    good_queue.appendleft(i)  # O(1): лише зміна вказівника
t_deque_queue = (time.perf_counter() - start) * 1000

print(f"  list.insert(0,...) × {n:,}: {t_list_queue:.2f} мс  → O(n²) загальна")
print(f"  deque.appendleft() × {n:,}: {t_deque_queue:.4f} мс  → O(n) загальна")
print(f"  Прискорення: {t_list_queue/t_deque_queue:.0f}x")

# -------------------------------------------
# АНТИПАТЕРН №3: Лінійний пошук vs bisect
# -------------------------------------------
print("\n=== Антипатерн №3: Пошук у відсортованому списку ===")

sorted_data = list(range(0, 1_000_000, 2))  # 500,000 парних чисел
target = 999_998  # Майже кінець списку — найгірший сценарій для лінійного пошуку

# O(n): Python ігнорує порядок і сканує лінійно
start = time.perf_counter()
for _ in range(1000):
    _ = target in sorted_data
t_linear = (time.perf_counter() - start) / 1000 * 1_000_000

# O(log n): bisect використовує бінарний пошук
start = time.perf_counter()
for _ in range(1000):
    idx = bisect.bisect_left(sorted_data, target)
    found = idx < len(sorted_data) and sorted_data[idx] == target
t_binary = (time.perf_counter() - start) / 1000 * 1_000_000

print(f"  'in' (лінійний пошук): {t_linear:.1f} мкс  → O(n): до 500,000 кроків")
print(f"  bisect (бінарний):     {t_binary:.2f} мкс  → O(log n): максимум ~19 кроків")
print(f"  Прискорення: {t_linear/t_binary:.0f}x")
````

## Частина 8 — Міні-вправи

Практика — єдиний спосіб засвоїти Big O. Вирішіть кожне завдання самостійно перед тим, як дивитись підказку.

---

### Вправа 1: Визначте Big O складність коду

Для кожного фрагменту коду визначте Big O складність та поясніть чому.

```python
# Фрагмент A
def mystery_a(data):
    return data[len(data) // 2]

# Фрагмент B
def mystery_b(data):
    result = []
    for item in data:
        result.append(item * 2)
    return result

# Фрагмент C
def mystery_c(data):
    seen = set()
    for item in data:
        if item in seen:
            return True
        seen.add(item)
    return False

# Фрагмент D
def mystery_d(data):
    n = len(data)
    for i in range(n):
        for j in range(i, n):
            if data[i] > data[j]:
                data[i], data[j] = data[j], data[i]
    return data
```

### Вправа 2: Виправте антипатерн

Наступний код знаходить першу дублікатовану цифру у списку. Він має O(n²) складність. Перепишіть його до O(n).

```python
# ПОВІЛЬНО: O(n²)
def has_duplicate_slow(numbers):
    for i in range(len(numbers)):
        for j in range(i + 1, len(numbers)):
            if numbers[i] == numbers[j]:
                return True
    return False
```

### Вправа 3: Проектування структури даних

Вам потрібно реалізувати систему прав доступу. У кожного користувача є список дозволених операцій. Система перевіряє дозволи **мільйони разів на секунду**.

```python
# Поточна реалізація:
user_permissions = ["read", "write", "delete", "admin"]

def can_do(user_permissions, action):
    return action in user_permissions  # Яка складність? Чи є кращий варіант?
```

Як переробити цю систему для максимальної продуктивності?

````python
# ===================================================
# РОЗВ'ЯЗКИ ВПРАВ (відкривайте тільки після спроби!)
# ===================================================

print("=" * 60)
print("ВПРАВА 1: Відповіді")
print("=" * 60)

print("""
Фрагмент A — O(1) Константний
  data[len(data) // 2] — знаходимо середній елемент за індексом.
  Незалежно від довжини списку — це завжди 1 операція.
  (Обчислення len() теж O(1) для list у Python)

Фрагмент B — O(n) Лінійний
  Один цикл по всіх n елементах.
  Кожен елемент обробляється рівно один раз.
  result.append() — O(1) амортизований, тому загалом O(n).

Фрагмент C — O(n) Лінійний (з set!)
  На перший погляд може здатися O(n²), але:
  - 'item in seen' — O(1), бо seen — це SET (хеш-таблиця)
  - 'seen.add(item)' — O(1)
  - Цикл: n ітерацій × O(1) всередині = O(n)
  Ключовий момент: якби 'seen' був списком — це стало б O(n²)!

Фрагмент D — O(n²) Квадратичний (Selection Sort)
  Вкладені цикли: зовнішній n ітерацій, внутрішній (n-i) ітерацій.
  Загальна кількість операцій: n*(n-1)/2 ≈ n²/2 → O(n²)
""")

print("=" * 60)
print("ВПРАВА 2: O(n) рішення для пошуку дублікатів")
print("=" * 60)

def has_duplicate_fast(numbers):
    """
    O(n) рішення: використовуємо set як 'пам'ять про побачені елементи'.
    
    Ідея: якщо елемент вже є в seen — це дублікат.
    seen будуємо по ходу → один прохід по списку.
    """
    seen = set()            # Порожня хеш-таблиця
    for num in numbers:     # O(n): один прохід
        if num in seen:     # O(1): хеш-пошук
            return True     # Знайшли дублікат!
        seen.add(num)       # O(1): додаємо до хеш-таблиці
    return False

# Перевірка
test1 = [1, 2, 3, 4, 5]
test2 = [1, 2, 3, 2, 5]
print(f"  has_duplicate_fast([1,2,3,4,5]): {has_duplicate_fast(test1)} (очікується False)")
print(f"  has_duplicate_fast([1,2,3,2,5]): {has_duplicate_fast(test2)} (очікується True)")
print("  Складність: O(n) — один прохід, O(1) пам'ять у кожній ітерації")

print()
print("=" * 60)
print("ВПРАВА 3: Оптимальна система прав доступу")
print("=" * 60)

# ПОГАНИЙ варіант: список → O(n) пошук
user_permissions_list = ["read", "write", "delete", "admin"]

# ХОРОШИЙ варіант: set → O(1) пошук
# Конвертуємо один раз при ініціалізації, потім мільйони перевірок за O(1)
user_permissions_set = set(user_permissions_list)

def can_do_bad(perms_list, action):
    """O(n): Python сканує весь список."""
    return action in perms_list  # Лінійний пошук!

def can_do_good(perms_set, action):
    """O(1): хеш-таблиця — миттєва відповідь."""
    return action in perms_set   # Хеш-пошук!

import time

# Симулюємо 1 мільйон перевірок прав доступу
start = time.perf_counter()
for _ in range(1_000_000):
    can_do_bad(user_permissions_list, "admin")
t_list = (time.perf_counter() - start) * 1000

start = time.perf_counter()
for _ in range(1_000_000):
    can_do_good(user_permissions_set, "admin")
t_set = (time.perf_counter() - start) * 1000

print(f"  list (O(n)) × 1,000,000 перевірок: {t_list:.0f} мс")
print(f"  set  (O(1)) × 1,000,000 перевірок: {t_set:.0f} мс")
print(f"  Прискорення: {t_list/t_set:.1f}x")
print("")
print("  Правильна архітектура:")
print("    permissions: set = frozenset(['read', 'write', 'admin'])")
print("    Використовуємо frozenset — незмінна (immutable) версія set,")
print("    яку можна використовувати як ключ словника або значення іншого set.")
````

### Головні ідеї уроку

**1. Big O — це мова масштабованості, не мікровимірювань.**
Вона відповідає на запитання: *«Якщо n зросте у 1000 разів, скільки разів зросте кількість операцій?»*. Секунди — характеристика заліза. Big O — характеристика алгоритму.

**2. Три правила аналізу:**
- Завжди розглядаємо **найгірший сценарій**
- **Ігноруємо константи**: O(5n) = O(n)
- **Вищий порядок домінує**: O(n² + n) = O(n²)

**3. Python: знай свої структури даних**

| Потреба | Структура | Чому |
|---------|-----------|------|
| Зберегти порядок, доступ за індексом | `list` | O(1) індексний доступ |
| Перевірити унікальність, O(1) lookup | `set` | Хеш-таблиця без значень |
| Знайти за ідентифікатором | `dict` | Хеш-таблиця з парами key-value |
| Черга (FIFO) або стек (LIFO) | `deque` | O(1) з обох кінців |
| Пошук у відсортованому списку | `bisect` | O(log n) бінарний пошук |

**4. Найпоширеніші пастки (антипатерни):**
- `item in list` всередині циклу → **O(n²)** → конвертуйте в `set` → **O(n)**
- `list.insert(0, item)` → **O(n)** → використовуйте `deque.appendleft()` → **O(1)**
- `item in sorted_list` → **O(n)** → використовуйте `bisect` → **O(log n)**

**5. Принцип Роба Пайка:** «Хитромудрі алгоритми повільно працюють при малих n, а n зазвичай маленьке». Не ускладнюйте без потреби. 15 елементів — не час для хеш-таблиць.

---

### Ієрархія оптимізації (від важливого до менш важливого)

```
1. Обирайте правильний алгоритм (O-велике класу)   ← НАЙВАЖЛИВІШЕ
2. Обирайте правильну структуру даних
3. Мінімізуйте операції всередині критичних циклів
4. Мікрооптимізації синтаксису                     ← НАЙМЕНШ ВАЖЛИВО
```

> Перехід від O(n²) до O(n) дасть вам **1,000,000x** приріст на великих даних.
> Переписування O(n²) на C дасть вам **5–10x** на тих самих великих даних.
> **Правильна структура даних вирішує архітектуру.**

### ❓ Питання 1

**«Яка різниця між `list` і `set` у Python, і коли ви використаєте кожну з них?»**

---

**Неправильна відповідь (Junior-пастка):**
> «Set не зберігає дублікати, тому я використовую його для унікальних елементів.»

Це правда, але неповна відповідь. Інтерв'юер хоче почути про продуктивність.

**Правильна відповідь:**

| Критерій | `list` | `set` |
|----------|--------|-------|
| Впорядкованість | Зберігає порядок вставки | Не гарантує порядок |
| Дублікати | Дозволяє | Забороняє |
| Доступ за індексом | `O(1)` — `my_list[5]` | Неможливо |
| Пошук (`in`) | `O(n)` — лінійне сканування | `O(1)` — хеш-таблиця |
| Вставка | `O(1)` append / `O(n)` insert | `O(1)` add |

**Архітектурний висновок:**
- `list` — коли потрібен **порядок** або **доступ за позицією**
- `set` — коли потрібна **перевірка існування** або **унікальність**, особливо всередині циклів

```python
# Класичний приклад де set рятує архітектуру:
# ПОГАНИЙ код: O(n²)
for user in users:
    if user.id in blocked_ids_list:  # O(n) пошук!
        skip(user)

# ХОРОШИЙ код: O(n)
blocked_set = set(blocked_ids_list)  # O(n) — один раз
for user in users:
    if user.id in blocked_set:       # O(1) кожен раз!
        skip(user)
```
___
**🎯 Як правильно сформулювати на співбесіді**

list використовується, коли важливий порядок або доступ за індексом.
set використовується, коли важлива швидка перевірка існування — O(1), що критично всередині циклів.
___

### ❓ Питання 2

**«Що таке амортизована складність? Чому `list.append()` вважається O(1)?»**

---

**Неправильна відповідь:**
> «Тому що Python завжди додає в кінець масиву за один крок.»

Це неточно. Часом `append` коштує O(n)!

**Правильна відповідь:**

Коли внутрішній масив пам'яті заповнюється, Python:
1. Виділяє новий масив розміром ×1.125 більший за поточний
2. Копіює всі наявні елементи в новий масив → це **O(n)**!
3. Додає новий елемент

Але це відбувається **логарифмічно рідко** (при 8, 16, 32, 64, 128... елементах).

**Амортизована вартість** = загальна вартість n операцій / n

Якщо ми виконаємо 1,000,000 append'ів:
- ~999,980 операцій: O(1) кожна
- ~20 операцій: O(n) (копіювання)
- Загальна вартість ≈ O(n log n) для серії
- **На одну операцію в середньому: O(1)**

```python
import sys

lst = []
for i in range(10):
    old_size = sys.getsizeof(lst)
    lst.append(i)
    new_size = sys.getsizeof(lst)
    if new_size != old_size:
        print(f"Реалокація пам'яті при n={len(lst)}: {old_size} → {new_size} байт")
```

> **Ключова ідея:** Амортизована складність — це *середня* вартість операції по всій серії викликів, а не вартість одного найгіршого виклику.
___
**🎯 Як відповісти**

append іноді виконується за O(n), коли відбувається реалокація пам’яті.
Але оскільки це трапляється рідко, середня вартість операції є O(1).
Це називається амортизована складність.

### ❓ Питання 3

**«Як би ви відрефакторили цю функцію для роботи з мільярдом записів? Поясніть ваші рішення через Big O.»**

```python
def get_active_premium_users(all_users, premium_user_ids, banned_user_ids):
    result = []
    for user in all_users:
        if user.id in premium_user_ids and user.id not in banned_user_ids:
            if user.last_login > THRESHOLD:
                result.append(user)
    return result
# Припустимо: all_users — list, premium_user_ids — list, banned_user_ids — list
```

---

**Аналіз проблем:**

1. `user.id in premium_user_ids` — O(p) де p = len(premium_user_ids)
2. `user.id not in banned_user_ids` — O(b) де b = len(banned_user_ids)
3. Зовнішній цикл — O(n) де n = len(all_users)

**Загальна складність: O(n × (p + b))** — при n=1B, p=1M, b=100K → трильйони операцій!

**Рішення Senior-розробника:**

```python
def get_active_premium_users_optimized(
    all_users,
    premium_user_ids,     # Передаємо ці дані один раз
    banned_user_ids
):
    # Крок 1: O(p + b) — одноразова побудова хеш-таблиць
    # Виносимо ЗА межі циклу — не повторюємо роботу n разів
    premium_set = frozenset(premium_user_ids)  # frozenset — незмінний, трохи швидший
    banned_set  = frozenset(banned_user_ids)

    # Крок 2: O(n) × O(1) × O(1) = O(n)
    # Генератор замість list для ледачої обробки — економить пам'ять
    return (
        user
        for user in all_users
        if user.id in premium_set     # O(1)
        and user.id not in banned_set # O(1)
        and user.last_login > THRESHOLD  # O(1)
    )

# Загальна складність: O(n + p + b)
# При n=1B, p=1M, b=100K → ~1.1 мільярда операцій (замість трильйонів)
```
---
**🎯 Як відповісти**

Проблема в тому, що перевірка in виконується на списках, що дає O(n) всередині циклу.
Я б перетворив ці списки у set, щоб отримати O(1) доступ і зменшити загальну складність до O(n).
___


**Додатково:**
- Якщо banned_set рідко змінюється → кешувати його (LRU cache / Redis)
- Якщо `all_users` не поміщається в RAM → використовувати генератор + потокову обробку
- Якщо потрібно паралельно → `multiprocessing.Pool` з розбивкою списку на чанки

> **Ключовий принцип:** Структури даних вирішують архітектуру. Ніяка мікрооптимізація не компенсує O(n²) алгоритм.

🧠 Головний меседж (що хоче почути інтерв’юер)
1. Я бачу складність алгоритму
2. Я розумію, де вона виникає
3. Я змінюю структуру даних, щоб її зменшити

````python
# ===================================================
# ПИТАННЯ НА СПІВБЕСІДІ: Демонстрація рішень
# ===================================================
import time
import sys
from dataclasses import dataclass
from datetime import datetime, timedelta

# -------------------------------------------
# Питання 1: list vs set — реальний вимір
# -------------------------------------------
print("=== Питання 1: list vs set для перевірки прав ===")

blocked_ids_list = list(range(100_000))  # 100,000 заблокованих ID (як список)
blocked_ids_set  = set(blocked_ids_list)  # Та сама інформація (як множина)

# Симулюємо 50,000 перевірок
test_ids = list(range(50_000, 150_000))  # Половина є в blocked, половина — ні

start = time.perf_counter()
count_bad = sum(1 for uid in test_ids if uid in blocked_ids_list)
t_list = (time.perf_counter() - start) * 1000

start = time.perf_counter()
count_good = sum(1 for uid in test_ids if uid in blocked_ids_set)
t_set = (time.perf_counter() - start) * 1000

print(f"  list lookup (O(n)): {t_list:.0f} мс | знайдено: {count_bad}")
print(f"  set lookup  (O(1)): {t_set:.1f} мс | знайдено: {count_good}")
print(f"  Прискорення: {t_list/t_set:.0f}x")

# -------------------------------------------
# Питання 2: Амортизований append — спостерігаємо реалокацію
# -------------------------------------------
print("\n=== Питання 2: Амортизований O(1) — реалокація пам'яті list ===")

lst = []
prev_size = sys.getsizeof(lst)
reallocations = []

for i in range(65):  # Перші 65 елементів
    lst.append(i)
    curr_size = sys.getsizeof(lst)
    if curr_size != prev_size:  # Пам'ять перевиділена!
        reallocations.append((len(lst), prev_size, curr_size))
        prev_size = curr_size

print(f"  {'Елементів':>10} | {'До (байт)':>10} | {'Після (байт)':>12} | Зростання")
print(f"  {'-'*50}")
for n_elem, before, after in reallocations:
    print(f"  {n_elem:>10} | {before:>10} | {after:>12} | +{after-before} байт")
print(f"  Усього реалокацій для 65 елементів: {len(reallocations)}")
print("  → Амортизований O(1): копіювання трапляється логарифмічно рідко")

# -------------------------------------------
# Питання 3: Рефакторинг функції фільтрації
# -------------------------------------------
print("\n=== Питання 3: Фільтрація активних premium-юзерів ===")

@dataclass
class User:
    id: int
    last_login: datetime

THRESHOLD = datetime.now() - timedelta(days=30)

# Генеруємо тестові дані
all_users       = [User(i, datetime.now() - timedelta(days=i % 60)) for i in range(50_000)]
premium_list    = list(range(0, 25_000))   # 25,000 premium-юзерів
banned_list     = list(range(10_000, 15_000))  # 5,000 заблокованих

# ПОВІЛЬНА версія: O(n * p + n * b)
def get_active_slow(users, premium_ids_list, banned_ids_list):
    result = []
    for user in users:
        if user.id in premium_ids_list and user.id not in banned_ids_list:
            if user.last_login > THRESHOLD:
                result.append(user)
    return result

# ШВИДКА версія: O(n + p + b)
def get_active_fast(users, premium_ids_list, banned_ids_list):
    premium_set = frozenset(premium_ids_list)  # O(p) — один раз
    banned_set  = frozenset(banned_ids_list)   # O(b) — один раз
    return [
        user for user in users
        if user.id in premium_set
        and user.id not in banned_set
        and user.last_login > THRESHOLD
    ]

start = time.perf_counter()
r1 = get_active_slow(all_users, premium_list, banned_list)
t_slow = (time.perf_counter() - start) * 1000

start = time.perf_counter()
r2 = get_active_fast(all_users, premium_list, banned_list)
t_fast = (time.perf_counter() - start) * 1000

print(f"  Повільна O(n*(p+b)): {t_slow:.0f} мс | знайдено: {len(r1)} юзерів")
print(f"  Швидка  O(n+p+b):   {t_fast:.1f} мс | знайдено: {len(r2)} юзерів")
print(f"  Прискорення: {t_slow/t_fast:.0f}x")
````

# ❌ Що таке “передчасна оптимізація”

Це коли ти робиш складніше рішення **без реальної причини**

---

## Приклад

```python
# Простий і нормальний код
if user.id in users_list:
    ...
```

А ти одразу робиш:

```python
users_set = set(users_list)

if user.id in users_set:
    ...
```

👉 хоча список має 10 елементів

---

## Проблема

* код став складнішим
* пам’ять використовується більше
* виграшу по швидкості нема

---

# ✅ Коли оптимізація потрібна

Тільки коли є **проблема в продуктивності**

---

## Правильний підхід

```text
1. Написати простий код
2. Запустити
3. Виміряти (time, profiling)
4. Знайти вузьке місце
5. Оптимізувати саме його
```

---

# 🧠 Як це виглядає у реальному житті

## Сценарій

Ти пишеш систему.

### Етап 1

```text
10 000 записів → все працює
```

👉 нічого не оптимізуєш

---

### Етап 2

```text
1 000 000 записів → починає гальмувати
```

👉 тепер:

* дивишся де проблема
* знаходиш O(n²)
* замінюєш на set → O(n)

---

# 💥 Чому Кнут сказав “навіть експертам — не оптимізувати”

Бо навіть досвідчені розробники:

* можуть помилитися
* оптимізувати не те місце
* ускладнити код без вигоди

---

# 🔥 Дуже важлива думка

> ❗ **Оптимізація — це реакція на проблему, а не стиль програмування**

---

# 🧩 Як це пов’язано з Big-O

Big-O — це інструмент, який каже:

```text
ось це буде проблемою на великих даних
```

Але:

* якщо даних мало → не проблема
* якщо даних багато → тоді дієш
